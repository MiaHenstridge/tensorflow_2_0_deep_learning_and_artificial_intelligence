## Start by doing the WRONG thing - trying to predict the price itself

```python
series = df['close'].values.reshape(-1, 1)
```

```python
# Normalize the data
scaler = StandardScaler()
scaler.fit(series[:len(series)//2])
series = scaler.transform(series).flatten()
```

```python
series.shape
```

```python
# build the dataset
T = 10
D = 1
X = []
Y = []

for t in range(len(series)-T):
    x = series[t:t+T]
    X.append(x)
    y = series[t+T]
    Y.append(y)
    
X = np.array(X).reshape(-1, T, 1)
Y = np.array(Y)
N = len(X)
```

```python
X.shape, Y.shape
```

```python
# try autoregressive RNN model
i = Input(shape=(T, 1))
x = LSTM(5)(i)
x = Dense(1)(x)

model = Model(i, x)
```

```python
model.compile(loss='mse',
             optimizer=Adam(learning_rate=0.1))
```

```python
r = model.fit(X[:-N//2], Y[:-N//2],
             epochs=80,
             validation_data=(X[-N//2:], Y[-N//2:]))
```

```python
# plot loss per iteration
plt.plot(r.history['loss'], label='loss')
plt.plot(r.history['val_loss'], label='val_loss')
plt.legend()
plt.show()
```

```python
# one-step forecast using true targets
outputs = model.predict(X)
print(outputs.shape)
predictions = outputs[:,0]
```

```python
plt.plot(Y, label='targets')
plt.plot(predictions, label='predictions')
plt.legend()
plt.show()
```

```python
# multi-step forecast
validation_target = Y[-N//2:]
validation_predictions = []

# first validation input
last_x = X[-N//2]        # 1D array of length T

while len(validation_predictions) < len(validation_target):
    p = model.predict(last_x.reshape(1, T, 1))[0, 0]
    # update the predictions list
    validation_predictions.append(p)
    # make new input
    last_x = np.roll(last_x, -1)
    last_x[-1] = p
```

```python
plt.plot(validation_target, label='forecast target')
plt.plot(validation_predictions, label='forecast prediction')
plt.legend()
```

```python
df.head()
```

```python
df['return'] = (df['close'] - df['prev_close'])/df['prev_close']
```

```python
df['return'].hist()
```

## predict with multi-dimensional time series

```python
input_data = df[['open', 'high', 'low', 'close', 'volume']].values
targets = df['return'].values
```

```python
input_data.shape, targets.shape
```

```python
# make the actual data which will go into the neural network
T = 10
D = input_data.shape[1]
N = len(input_data) - T
```

```python
N, T, D
```

```python
# normalize inputs
Ntrain = len(input_data) * 2 // 3
scaler = StandardScaler()

scaler.fit(input_data[:Ntrain + T])

input_data = scaler.transform(input_data)
```

```python
input_data.shape
```

```python
# set up X_train and Y_train
X_train = np.zeros((Ntrain, T, D))
Y_train = np.zeros(Ntrain)
```

```python
for t in range(Ntrain):
    X_train[t, :, :] = input_data[t:t+T]
    Y_train[t] = (targets[t+T] > 0)
```

```python
# set up X_test and Y_test
X_test = np.zeros((N - Ntrain, T, D))
Y_test = np.zeros(N - Ntrain)

for u in range(N - Ntrain):
    X_test[u, :, :] = input_data[u:u+T]
    Y_test[u] = (targets[u+T] > 0)
    
```

```python
# make the RNN
i = Input(shape=(T, D))
x = LSTM(50)(i)
x = Dense(1, activation='sigmoid')(x)

model = Model(i, x)
```

```python
model.compile(loss='binary_crossentropy',
             optimizer=Adam(learning_rate=0.001),
             metrics=['accuracy'])
```

```python
r = model.fit(X_train, Y_train,
             batch_size=32,
             validation_data=(X_test, Y_test),
             epochs=300)
```

```python
plt.plot(r.history['loss'], label='loss')
plt.plot(r.history['val_loss'], label='val_loss')
plt.legend()
plt.show()
```

```python
# Plot accuracy per iteration
plt.plot(r.history['accuracy'], label='accuracy')
plt.plot(r.history['val_accuracy'], label='val_accuracy')
plt.legend()
plt.show()
```
