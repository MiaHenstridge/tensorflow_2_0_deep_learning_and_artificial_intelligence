```python
# L(w) = w**2

w = tf.Variable(5.)

def get_loss(w):
    return w**2

def get_grad(w):
    with tf.GradientTape() as tape:
        L = get_loss(w)
    g = tape.gradient(L, w)
    return g

optimizer = tf.keras.optimizers.SGD(learning_rate=0.1)

losses = []

for i in range(50):
    g = get_grad(w)
    optimizer.apply_gradients(zip([g], [w]))
    losses.append(get_loss(w))
```

```python
# plot losses
import matplotlib.pyplot as plt
plt.plot(losses)

print(f"final loss: {get_loss(w)}")
```

```python
# full computation
w = tf.Variable(5.)
losses2 = []

for i in range(50):
    w = w - 0.1 * 2 * w
    losses2.append(w**2)
```

```python
# plot both losses on the same plot to compare results
# they should overlap
plt.plot(losses, label='losses tf')
plt.plot(losses2, label='losses manual')
plt.legend()
```
